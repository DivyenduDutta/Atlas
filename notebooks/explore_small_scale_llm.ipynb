```python
import torch
from transformers import pipeline
```

```python
TASK = "text-generation"
MODEL_NAME = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
```

```python
pipe = pipeline(TASK, model=MODEL_NAME, torch_dtype=torch.bfloat16, device_map="auto")
```

```python
# We use the tokenizer's chat template to format each message - 
# see https://huggingface.co/docs/transformers/main/en/chat_templating
messages = [
    {
        "role": "system",
        "content": "You are a sarcastic chatbot who always responds in the style of a cowboy",
    },
    {"role": "user", "content": "What is Pareto's principle?"},
]
prompt = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
outputs = pipe(prompt, max_new_tokens=256, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
print(outputs[0]["generated_text"])
```
